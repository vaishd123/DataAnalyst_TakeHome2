```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('email_data.csv')
data
```

```python
def dateextract(DateAndTime):
    return DateAndTime.split()[0]
```

```python
data['Date']=data['EmailSentDateTimeUtc'].apply(dateextract)
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['year'] = pd.DatetimeIndex(data['Date']).year
data['month'] = pd.DatetimeIndex(data['Date']).month
data['year-month'] = data['Date'].apply(lambda x: str(x.year) + ' ' + str(x.month))
```

```python
result = data.groupby(['year-month']).agg({"UserId": "nunique"})
```

```python
result
```

```python
sns.barplot(x = result.index, y = result.UserId)
sns.set(rc={'figure.figsize':(15.7,8.27)})
plt.xlabel("Names")
plt.xticks(rotation=45)


plt.ylabel("Age of the person")
ax = plt.axes()

plt.show()
```
